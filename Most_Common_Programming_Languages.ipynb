### Most Popular Programming Languages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import HowToBreakIntoTheField as t
%matplotlib inline

df = pd.read_csv('./survey_results_public.csv')
schema = pd.read_csv('./survey_results_schema.csv')
df.head()
```

Provide a pandas series of the different **WantWorkLanguage** status values in the dataset.  Store this pandas series in **lang_vals**.

```python
lang_vals = df.WantWorkLanguage.value_counts()#Provide a pandas series of the counts for each WantWorkLanguage

lang_vals # assure this looks right
```

```python
# The below should be a bar chart of the proportion of languages in your lang_vals
# if it is set up correctly.

(lang_vals/df.shape[0]).plot(kind="bar");
plt.title("WantWorkLanguage");
```

We must clean the data. The plot above shows you how it looks like without cleaning.

```python
possible_vals = ["JavaScript","Python","Java","C#","SQL","TypeScript","Swift","PHP","Go","C++","Scala","Smalltalk",
                 "Ruby","C","F#","Erlang","Rust","Haskell","Objective-C","Elixir","R","Groovy","Julia","Perl",
                 "Assembly","Lua","Dart","CoffeeScript","Matlab","VBA","Clojure","VB.NET","Common Lisp",
                 "Visual Basic 6","Hack"]

def clean_and_plot(df, title='Most Popular Programming Languages Among Developers ', plot=True):

    study = df['WantWorkLanguage'].value_counts().reset_index()
    study.rename(columns={'index': 'Programming Language', 'WantWorkLanguage': 'count'}, inplace=True)
    study_df = t.total_count(study, 'Programming Language', 'count', possible_vals)

    study_df.set_index('Programming Language', inplace=True)
    if plot:
        (study_df/study_df.sum()).plot(kind='bar', legend=None);
        plt.title(title);
        plt.show()
    props_study_df = study_df/study_df.sum()
    return props_study_df
    
props_df = clean_and_plot(df)
```

We applied the same for "Have Worked Language" too see results from different perspective.

```python
possible_vals = ["JavaScript","Python","Java","C#","SQL","TypeScript","Swift","PHP","Go","C++","Scala","Smalltalk",
                 "Ruby","C","F#","Erlang","Rust","Haskell","Objective-C","Elixir","R","Groovy","Julia","Perl",
                 "Assembly","Lua","Dart","CoffeeScript","Matlab","VBA","Clojure","VB.NET","Common Lisp",
                 "Visual Basic 6","Hack"]

def clean_and_plot(df, title='Have Worked Language', plot=True):

    study = df['HaveWorkedLanguage'].value_counts().reset_index()
    study.rename(columns={'index': 'language', 'HaveWorkedLanguage': 'count'}, inplace=True)
    study_df = t.total_count(study, 'language', 'count', possible_vals)

    study_df.set_index('language', inplace=True)
    if plot:
        (study_df/study_df.sum()).plot(kind='bar', legend=None);
        plt.title(title);
        plt.show()
    props_study_df = study_df/study_df.sum()
    return props_study_df
    
props_df = clean_and_plot(df)
```
